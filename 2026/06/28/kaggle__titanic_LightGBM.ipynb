```python
import pandas as pd
from torch.utils.data import DataLoader, TensorDataset, random_split
import torch

test_csv = pd.read_csv(path+'/test.csv')
train_csv = pd.read_csv(path+'/train.csv')

print(f"shape:{test_csv.shape, train_csv.shape}")

train_csv
```

```python

csv = pd.concat([train_csv, test_csv], axis=0, ignore_index=True)
csv.ffill(inplace=True)
csv.bfill(inplace=True)


names = csv['Name']
cabin = csv['Cabin']
cabin_letter = cabin.astype(str).str[0]

csv = csv.drop(columns='Name', axis=1)
csv = csv.drop(columns='Cabin', axis=1)



csv['Married'] = names.str.contains('Mrs.', na=False).astype(int)
csv['Master'] = names.str.contains('Master.', na=False).astype(int)
csv['Cabin_col'] = cabin_letter.astype('category').cat.codes

csv = csv.drop(columns='Ticket', axis=1)

csv = pd.get_dummies(csv, dtype=int, drop_first=True)

test_csv = csv.iloc[train_csv.shape[0]:]
train_csv = csv.iloc[:train_csv.shape[0]]

csv
```

```python
print(f"shape:{test_csv.shape, train_csv.shape}")
```

```python
def to_dataset(csv):
    X = torch.tensor(csv.iloc[:, 2:].values, dtype=torch.float32)
    y = torch.tensor(csv.iloc[:, 1].values, dtype=torch.long)
    return TensorDataset(X, y)

dataset = to_dataset(train_csv)

train_size = int(0.8 * len(dataset))
valid_size = len(dataset) - train_size

train_dataset, valid_dataset = random_split(
    dataset, 
    [train_size, valid_size],
    generator=torch.Generator().manual_seed(42)
)
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
import lightgbm as lgb
import numpy as np
import torch

# 1. 創立紀錄容器 (LightGBM 也有類似訓練/驗證 Loss 的指標：評價函數得分)
total_loss = {"valid": [], "train": []}

# 2. 【核心修改】將 PyTorch DataLoader 轉換為 LightGBM 格式
# LightGBM 需要一口氣吃完整個矩陣，所以我們把 DataLoader 的資料拉出來並轉成 NumPy
def dataloader_to_numpy(dataloader):
    all_features, all_labels = [], []
    for features, labels in dataloader:
        # 將 PyTorch 張量扁平化並轉成 NumPy
        all_features.append(features.view(features.size(0), -1).numpy())
        all_labels.append(labels.numpy())
    return np.vstack(all_features), np.concatenate(all_labels)

train_iter = DataLoader(train_dataset, 128)
valid_iter = DataLoader(valid_dataset, 128)

# 假設您原本的 train_iter 和 valid_iter 已經建立好
X_train, y_train = dataloader_to_numpy(train_iter)
X_val, y_val = dataloader_to_numpy(valid_iter)

# 3. 建立 LightGBM 專屬的 Dataset 格式
train_data = lgb.Dataset(X_train, label=y_train)
valid_data = lgb.Dataset(X_val, label=y_val, reference=train_data)

# 4. 設定參數 (二分類任務，使用多標籤交叉熵 loss)
params = {
    "objective": "binary",  # 二分類
    "metric": "binary_logloss",  # 類似 PyTorch 的 CrossEntropyLoss
    "boosting_type": "gbdt",
    "learning_rate": 0.05,
    "verbosity": -1,  # 減少不必要的日誌輸出
}

# 5. 訓練模型 (一鍵完成，完全不需要自己寫 for epoch 迴圈！)
# eval_hist 字典會自動幫您存下訓練與驗證過程中的每一輪的 Loss 變化
eval_hist = {}
model = lgb.train(
    params,
    train_data,
    num_boost_round=100,  # 迭代 100 棵樹 (類似跑 100 次迭代)
    valid_sets=[train_data, valid_data],
    valid_names=["train", "valid"],
    callbacks=[lgb.record_evaluation(eval_hist)],
)

# 6. 把 Loss 紀錄塞回您原先的 total_loss 變數中，完全對接！
total_loss["train"] = eval_hist["train"]["binary_logloss"]
total_loss["valid"] = eval_hist["valid"]["binary_logloss"]

print(
    f"最終訓練完成！Train Loss: {total_loss['train'][-1]:.4f} | Valid Loss: {total_loss['valid'][-1]:.4f}"
)
```

```python
import pandas as pd

# 1. LightGBM 不需要 Tensor，直接將 Pandas DataFrame 轉成 NumPy 陣列即可
# 完全不麻煩，不用管 device（LightGBM 預設使用 CPU，速度極快）
X_test = test_csv.iloc[:, 2:].values

# 2. 【核心修改】使用 LightGBM 的預測語法
# LightGBM 沒有 eval() 或 no_grad()，直接呼叫 predict()
y_pred_prob = model.predict(X_test)

# 3. 將預測出的「生存機率」轉換為「0 或 1」的類別標籤
# 只要機率 >= 0.5 就判定為生還 (1)，否則為死亡 (0)
pred = (y_pred_prob >= 0.5).astype(int)

# 4. 產生 Submission 檔案 (保持您原本的邏輯)
submission = pd.DataFrame(
    {
        "PassengerId": pd.read_csv(path + "/gender_submission.csv").iloc[:, 0],
        "Survived": pred,
    }
)

submission_name = (
    "titanic_submission__drop-Ticket_change-Name-Cabin-LightGBM.csv"
)
submission.to_csv(submission_name, index=False)

print(f"預測完成！Submission 已儲存至 {submission_name}")
```
